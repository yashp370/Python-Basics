```python
# for the grid 
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image  # Correct import statement

# Path to the image
fname = r'iron.jpg'  # (for png image)

# Open image using PIL and convert to grayscale
img = Image.open(fname).convert('L')  # Use a different variable name to avoid conflicts

# Display image in grayscale
plt.imshow(img, cmap='RdPu')
plt.title('The Iron Man')
plt.xlabel('Length')
plt.ylabel('Height')
plt.grid()
# Commenting out the axis off line to ensure labels are visible
# plt.axis('off')

plt.show()
```

```python
x = [10,20,30,40]
y = ['english','hindi','math','science']
plt.pie(x , labels = y)
plt.show()
```

```python
#color
x = [10,20,30,40]
y = ['english','hindi','math','science']
c = ['y','m','b','c']
plt.pie(x , labels = y,colors = c)
plt.show()
```

```python
x = [10,20,30,40]
y = ['english','hindi','math','science']
c = ['y','m','b','c']
plt.pie(x , labels = y,colors = c)
plt.legend()
plt.show()
```

```python
# for the grid 
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image  # Correct import statement

# Path to the image
fname = r'iron.jpg'  # (for png image)

# Open image using PIL and convert to grayscale
img = Image.open(fname).convert('L')  # Use a different variable name to avoid conflicts

# Display image in grayscale
plt.imshow(img, cmap='RdPu')
plt.title('The Iron Man')
plt.xlabel('Length')
plt.ylabel('Height')
plt.grid()
plt.savefig(fname)
# Commenting out the axis off line to ensure labels are visible
# plt.axis('off')

plt.show()
```
