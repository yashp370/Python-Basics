```python
# bar plot in seaborn 

import seaborn as sns
import matplotlib.pyplot as ply
import pandas as pd
```

```python
sns.load_dataset("penguins")
```

```python
df1 = sns.load_dataset("penguins")
```

```python
df1
```

```python
df1['island']
```

```python
# different way ..
sns.barplot(x = df1.island , y = df1.bill_length_mm)
plt.show()
```

```python
sns.barplot(x = 'island', y = 'bill_length_mm' , data = df1)
plt.show()
```

```python
sns.barplot(x = 'island', y = 'bill_length_mm' , data = df1 , hue='sex')
plt.show()
```

```python
# order parameter (to change the x axis ..)
order1 = ['Dream' , 'Torgersen' , 'Biscoe']

sns.barplot(x = 'island', y = 'bill_length_mm' , data = df1 , hue = 'sex' ,order = order1 )
plt.show()
```

```python
# hue_order (to change the male and female..)

order1 = ['Dream' , 'Torgersen' , 'Biscoe']

sns.barplot(x = 'island', y = 'bill_length_mm' , data = df1 , hue = 'sex' ,order = order1  , hue_order=['Female' , 'Male'])
plt.show()
```

```python
# how too use colors ...

order1 = ['Dream' , 'Torgersen' , 'Biscoe']
sns.barplot(x = 'island' , y = 'bill_length_mm' , data = df1,color ='black')
plt.show()
```

```python
sns.barplot(x = 'island' , y = 'bill_length_mm', data = df1 ,hue='sex' , palette = 'twilight_shifted')
plt.show()
```

```python
sns.barplot(x = 'island' , y = 'bill_length_mm', data = df1 ,hue='sex' , palette = 'bright')
plt.show()
```

```python
# Saturation parameter ...
#(to change the color off a light or dark) 
#form (0 to 1) we can give saturation color


sns.barplot(x = 'island' , y = 'bill_length_mm', data = df1 ,hue='sex' , palette = 'bright' , saturation = 0.2)
plt.show()
```

                 histogram uses the   
                 Kde paramenters (Kernel Density Estimate)
    Kernel Density Estimate (KDE) is a statistical technique to estimate the probability density function of a continuous random variable.
     In the context of histograms, it provides a smoothed curve that represents the distribution of data.

```python
# using kde
sns.displot(df1['flipper_length_mm'] , bins = [170,180,190,200,210,220,230,240] , kde = True )
plt.show()
```

```python
sns.displot(df1['flipper_length_mm'] , bins = [170,180,190,200,210,220,230,240] , kde = False )
plt.show()
```

```python
sns.displot(df1['flipper_length_mm'] , bins = [170,180,190,200,210,220,230,240] , kde = True , color='green' )
plt.show()
```
