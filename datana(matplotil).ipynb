        Bar Graph

```python
x = [1,2,3,4,5]
y = [6,7,8,9,10]

plt.bar(x,y)
plt.show()
```

```python
# how to change color of bar

x = [1,2,3,4,5]
y = [6,7,8,9,10]
c = ['r' , 'g' , 'b' , 'y' ,'c'] 

plt.bar(x,y,color=c)

plt.show()
```

```python
# how i can horizontal the bar garph
# by using arrays
import numpy as np
```

```python
x = np.array(['A','B','C','D'])
y = np.array([3,5,8,15])

plt.bar(x,y)
plt.show()
```

```python
x = np.array(['A','B','C','D'])
y = np.array([3,5,8,15])

plt.barh(x,y)
plt.show()
```

```python
x = np.array(['A','B','C','D'])
y = np.array([3,5,8,15])

plt.barh(x,y , color = 'r')
plt.show()
```

```python
x = np.array(['A','B','C','D'])
y = np.array([3,5,8,15])

plt.bar(x,y,color = 'red')
plt.show()
```

```python
x = np.array(['A','B','C','D'])
y = np.array([3,5,8,15])

plt.bar(x,y,color = '#4CAF50')
plt.show()
```

```python
# width can be  moot or ptala 

x = np.array(['A','B','C','D'])
y = np.array([3,5,8,15])

plt.bar(x,y,color = 'c', width = 0.1)
plt.show()
```

```python
x = np.array(['A','B','C','D'])
y = np.array([3,5,8,15])

plt.bar(x,y,color = 'c', width = 1)
plt.show()
```

```python
# hight of the graph

x = np.array(['A','B','C','D'])
y = np.array([3,5,8,15])

plt.barh(x,y,color = 'c',  height = 0.2)
plt.show()
```

        Scatter plot

```python
x = [1,2,3,4,5,6]
y = [4,2,3,5,1,8]

plt.scatter(x,y)
plt.show()
```

```python
    # how to do title functio

x = [1,2,3,4,5,6]
y = [4,2,3,5,1,8]

plt.title('The title of the function is the Scatter Plot')

plt.scatter(x,y)
plt.show()
```

```python
# how to use label

x = [1,2,3,4,5,6]
y = [4,2,3,5,1,8]

plt.title('The title of the function is the Scatter Plot')

plt.xlabel("month")
plt.ylabel('days')

plt.scatter(x,y)
plt.show()
```

```python

# how to put color
x = [1,2,3,4,5,6]
y = [4,2,3,5,1,8]

plt.title('The title of the function is the Scatter Plot')

plt.xlabel("month")
plt.ylabel('days')

plt.scatter(x,y , color = 'r')
plt.show()
```

```python

x = [1,2,3,4,5,6]
y = [4,2,3,5,1,8]

plt.title('The title of the function is the Scatter Plot')

plt.xlabel("month")
plt.ylabel('days')

c =['r','g','b','c','y','m']

plt.scatter(x,y , color = c)
plt.show()
```

```python
# how to change size


x = [1,2,3,4,5,6]
y = [4,2,3,5,1,8]

plt.title('The title of the function is the Scatter Plot')

plt.xlabel("month")
plt.ylabel('days')

c =['r','g','b','c','y','m']

plt.scatter(x,y , color = c , s = 100)
plt.show()
```
